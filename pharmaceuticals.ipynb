```python
import os
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# csv_path=os.path.join('raw_data','clinicaltrial_data.csv')
# with open(csv_path, newline='') as csv_file:
#     data=csv.reader(csv_file,delimiter=',')
    
csv_path=os.path.join('raw_data','clinicaltrial_data.csv')
df=pd.DataFrame.from_csv(csv_path)

df.head()
```

```python
csv_path=os.path.join('raw_data','mouse_drug_data.csv')
df2=pd.DataFrame.from_csv(csv_path)

df2.head()
```

```python
df=df.reset_index()
df2=df2.reset_index()
```

```python
data=pd.merge(df,df2,on='Mouse ID',how='outer')
full_data=data.to_csv("raw_data/data_F.csv")
data.head()
```

```python
#Just to get an idea of the main dataset
grouped_data=data.groupby('Drug')

grouped_data.count()
```

```python
#getting all drug names in the grouped data above
grouped_data.groups.keys()

#grouping the data by drug, turning into a dataframe and sorting by time
grouped_df=data.groupby('Drug').apply( pd.DataFrame.sort_values, 'Timepoint' )

#select columns to show from grouped_df dataframe
grouped_df=grouped_df[['Mouse ID','Timepoint','Tumor Volume (mm3)','Metastatic Sites']]

#manually getting each of the drug dataframes seperately:

# capo=grouped_df.loc[['Capomulin'],['Timepoint','Tumor Volume (mm3)']]
# cefta=grouped_df.loc[['Ceftamin'],:]
# infu=grouped_df.loc[['Infubinol'],:]
# keta=grouped_df.loc[['Ketapril'],:]
# naft=grouped_df.loc[['Naftisol'],:]
# place=grouped_df.loc[['Placebo'],:]
# prop=grouped_df.loc[['Propriva'],:]
# rami=grouped_df.loc[['Ramicane'],:]
# stel=grouped_df.loc[['Stelasyn'],:]
# zoni=grouped_df.loc[['Zoniferol'],:]

#get the dataframe for each of the drugs via a loop   
for i in grouped_data.groups.keys():
    i=grouped_df.loc[[str(i)],:]
    print(i)
    
```

```python
##TUMOR RESPONSE TIME TO DIFFERENT TREATMENTS
```

```python
# Multiple rat IDs at same timepoints on the same drug
#Group by drug and timepoint and take the average of tumor volume
tumor_avg= data.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]
tumor_avg_df=pd.DataFrame(tumor_avg)

tumor_avg_df.head()
tumor_avg_df.tail()
```

```python
#calculate standard error of the mean, SEM, for tumor volume
tumor_sem= data.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]
tumor_sem_df = pd.DataFrame(tumor_sem)

tumor_sem_df=tumor_sem_df.rename(columns={"Tumor Volume (mm3)":'Tumor SEM'})
tumor_sem_df.head()
```

```python
# reset index of average and sem dataframes to pivot the dataframes and get drugs listed in columns instead of rows
tumor_avg_df = tumor_avg_df.reset_index()
tumor_sem_df = tumor_sem_df.reset_index()

tumor_avg_pivoted_df = tumor_avg_df.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]
tumor_sem_pivoted_df = tumor_sem_df.pivot(index="Timepoint", columns="Drug")["Tumor SEM"]

tumor_sem_pivoted_df.head()
```

```python
tumor_avg_pivoted_df.head()
```

```python

#create scatter plots that include the sem bars for the top 4 drugs
plt.errorbar(tumor_avg_pivoted_df.index, tumor_avg_pivoted_df["Capomulin"], yerr=tumor_sem_pivoted_df["Capomulin"], color="b",linestyle="-", linewidth=0.60,marker="s",markersize=3)
plt.errorbar(tumor_avg_pivoted_df.index, tumor_avg_pivoted_df["Ketapril"], yerr=tumor_sem_pivoted_df["Ketapril"], color="r", linestyle="-", linewidth=0.60,marker="^", markersize=3)
plt.errorbar(tumor_avg_pivoted_df.index, tumor_avg_pivoted_df["Infubinol"], yerr=tumor_sem_pivoted_df["Infubinol"], color="c", linestyle="-", linewidth=0.60, marker="+", markersize=3)
plt.errorbar(tumor_avg_pivoted_df.index, tumor_avg_pivoted_df["Placebo"], yerr=tumor_sem_pivoted_df["Placebo"], color="m", linestyle="-", linewidth=0.60, marker="h", markersize=3)

plt.title("Tumor Response to Treatment by Volume")
plt.ylabel("Tumor Volume in mm3")
plt.xlabel("Time in Days")
plt.legend(loc="best", fontsize="small")

plt.savefig("Tumor_volume_response_1.png")
plt.show()
```

```python
##METASTATIC SITES NUMBER IN RESPONSE TO TREATMENT
```

```python
# repeat step 1 of analysis but this time with the metastatic sites column of data and convert to dataframe
sites_avg = data.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]
sites_avg_df = pd.DataFrame(sites_avg)

sites_avg_df.head()

#compute the standard error of the mean of metastatic sites
sites_sem = data.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]
sites_sem_df = pd.DataFrame(sites_sem)
sites_sem_df=sites_sem_df.rename(columns={"Metastatic Sites":'Sites SEM'})


sites_sem_df.head()
```

```python
#as in part one, reset index of average and sem dataframes to pivot the dataframes and get drugs listed in columns instead of rows
sites_avg_df = sites_avg_df.reset_index()
sites_sem_df = sites_sem_df.reset_index()

sites_avg_pivoted_df= sites_avg_df.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]
sites_sem_pivoted_df =sites_sem_df.pivot(index="Timepoint", columns="Drug")["Sites SEM"]

sites_avg_pivoted_df.head()
sites_sem_pivoted_df.head()
```

```python
#as in part 1, create scatter plots that include the sem bars for the top 4 drugs
plt.errorbar(sites_avg_pivoted_df.index, sites_avg_pivoted_df["Capomulin"], yerr=sites_sem_pivoted_df["Capomulin"], color="b",linestyle="-", linewidth=0.60,marker="s",markersize=3)
plt.errorbar(sites_avg_pivoted_df.index, sites_avg_pivoted_df["Ketapril"], yerr=sites_sem_pivoted_df["Ketapril"], color="r", linestyle="-", linewidth=0.60,marker="^", markersize=3)
plt.errorbar(sites_avg_pivoted_df.index, sites_avg_pivoted_df["Infubinol"], yerr=sites_sem_pivoted_df["Infubinol"], color="c", linestyle="-", linewidth=0.60, marker="+", markersize=3)
plt.errorbar(sites_avg_pivoted_df.index, sites_avg_pivoted_df["Placebo"], yerr=sites_sem_pivoted_df["Placebo"], color="m", linestyle="-", linewidth=0.60, marker="h", markersize=3)

plt.title("Change in Metastatic Sites During Treatment")
plt.ylabel("Number of Metastatic Sites")
plt.xlabel("Treatment Time in Days)")
plt.legend(loc="best", fontsize="small")

plt.savefig("Metastatic_Site_Change.png")
plt.show()
```

```python
#ANALYSIS OF NUMBER OF MICE STILL ALIVE DURING TREATMENT: SURVIVAL RATE
```

```python
#Mouse count from beginning to end of treatment time with each of the drugs
mouse_count = data.groupby(["Drug", "Timepoint"]).count()["Tumor Volume (mm3)"]

mouse_count_df = pd.DataFrame(mouse_count)
mouse_count_df=mouse_count_df.rename(columns={'Tumor Volume (mm3)':'Survived Mouse Count'})
mouse_count_df.head()
```

```python
#reset index of mouse count dataframe to pivot the dataframe and get drugs listed in columns
mouse_count_df = mouse_count_df.reset_index()
count_pivoted_df = mouse_count_df.pivot(index="Timepoint", columns="Drug")["Survived Mouse Count"]

count_pivoted_df.head()
```

```python
#calculate survival rates
survival_rate_capo=(count_pivoted_df['Capomulin']/count_pivoted_df['Capomulin'].max())*100
survival_rate_keta=(count_pivoted_df['Ketapril']/count_pivoted_df['Ketapril'].max())*100
survival_rate_infu=(count_pivoted_df['Infubinol']/count_pivoted_df['Infubinol'].max())*100
survival_rate_pla=(count_pivoted_df['Placebo']/count_pivoted_df['Placebo'].max())*100

#plot survival rates versus time
plt.plot(survival_rate_capo, "o", linestyle="--", markersize=3, linewidth=0.60)
plt.plot(survival_rate_keta, "^", linestyle="--", markersize=3, linewidth=0.60)
plt.plot(survival_rate_infu, "s", linestyle="--", markersize=3, linewidth=0.60)
plt.plot(survival_rate_pla, "d", linestyle="--", markersize=3, linewidth=0.60)
plt.title("Mouse Survival through Time During Drug Treatment")
plt.ylabel("Mice Survival Rate")
plt.xlabel("Time in Days")
plt.legend(loc="best", fontsize="small")
plt.grid()

plt.savefig("Survival.png")
plt.show()

```

```python
##CREATING BAR GRAPH THAT COMPARES TOTAL % TUMOR VOL. CHANGE FOR EACH DRUG IN THE FULL 45 DAYS DURATION
```

```python
#the percentage change in tumor size
avg_delta_tumor =  ((tumor_avg_pivoted_df.iloc[-1] - tumor_avg_pivoted_df.iloc[0]) / tumor_avg_pivoted_df.iloc[0])*100
# sem_delta_tumor =  ((tumor_sem_pivoted_df.iloc[-1] - tumor_sem_pivoted_df.iloc[0]) / tumor_sem_pivoted_df.iloc[0])*100
avg_delta_tumor
# sem_delta_tumor
```

```python
tumor_changes = (avg_delta_tumor["Capomulin"],avg_delta_tumor["Infubinol"], avg_delta_tumor["Ketapril"], avg_delta_tumor["Placebo"])

# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
indices = np.arange(len(tumor_changes))  
width = 1
rectsPass = ax.bar(indices[0], tumor_changes[0], width, color='green')
rectsFail = ax.bar(indices[1:], tumor_changes[1:], width, color='red')

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel('Percentage Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Days')
ax.set_xticks(indices)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.set_autoscaley_on(False)
ax.set_ylim([-30,70])
ax.grid(True)

# Use functions to label the percentages of changes
def autolabelFail(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 3,
                '%d%%' % int(height),
                ha='center', va='bottom', color="white")

def autolabelPass(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., -8,
                '-%d%% ' % int(height),
                ha='center', va='bottom', color="white")

# Call functions to implement the function calls
autolabelPass(rectsPass)
autolabelFail(rectsFail)

# Save the Figure
fig.savefig("bar_chart.png")

# Show the Figure
fig.show()
```
